Since the research area for this study is Leeds City and the target group is students, an initial data cleaning step is necessary to remove irrelevant parts of the dataset to make the subsequent data processing more efficient. In the economic activity status data, only two variables needed to be retained: â€œEconomically active and a full-time studentâ€ and â€œEconomically inactive: Student.â€ These two variables represent the number of students with different economic activity statuses within each LSOA area. In addition, data for the city of Leeds was selected using the â€œLower Layer Super Output Areasâ€ field in the dataset. Since the student economic activity data is from 2021, only the â€œchn2021â€ field was retained from the residential mobility index dataset to ensure consistency in the research data. This variable: â€œchn2021,â€ represents the residential mobility index for each LSOA in 2021.

Real-world data is often inconsistent, duplicated, inaccurate, incomplete, or stale. Such data usually leads to misleading or biased analysis results in research [2]. There is a significant disparity between the maximum and minimum values in the economic activity status data according to data exploration. To further examine the data distribution and identify outliers, a strip plot generated by seaborn.catplot() was used for visualization. This function can display the relationship between numerical values and one or more categorical variables [3]. Through the visualization of student numbers across different economic statuses, a notably large outlier was found in both the economically active and inactive data categories. Meanwhile, further data exploration showed that both outliers appear in the same LSOA: E01011678. According to information found through web searches, this area is located near some universities and contains a large number of student accommodations. Therefore, these two values are considered valid values and were retained.
```python
import geopandas as gpd
import matplotlib.colors as colors
import matplotlib.patches as mpatches
import contextily as ctx
from shapely import wkt
from geo_northarrow import add_north_arrow
from matplotlib_scalebar.scalebar import ScaleBar
```

```python
inactive_student=data_m[data_m['Economic activity status'] == 'Economically inactive: Student']
```

```python
leeds_shp =gpd.read_file('https://github.com/FrancescaPontin/GEOG5990M/raw/refs/heads/main/data/week_20/Leeds.geojson')
```

```python
leeds_shp.head()
```

```python
merged_data = leeds_shp.merge(inactive_student, how='left',left_on='LSOA21CD',right_on='Lower layer Super Output Areas Code')
```

```python
merged_data.head()
```

```python
merged_data.isnull().sum() #check for missing values
```

```python
merged_data.loc[merged_data['Observation'].isna()] #check for missing values location
```

```python
new_data = merged_data

# Fill missing values in the Observation column with zero
new_data['Observation'] = new_data['Observation'].fillna(0)

# Prepare for bivariate classification - divide both variables into tertiles (Low, Medium, High)
# For students: compute tertiles only using non-zero values
nonzero_students = new_data[new_data['Observation'] > 0]['Observation']
student_tertiles = [
    nonzero_students.quantile(0.33),
    nonzero_students.quantile(0.67)
]

# Compute tertiles for Residential Mobility Index (chn2021)
mobility_tertiles = [
    new_data['chn2021'].quantile(0.33),
    new_data['chn2021'].quantile(0.67)
]

# Define classification functions for each variable
def classify_students(value, tertiles):
    if value == 0:
        return 0  # No students
    elif value <= tertiles[0]:
        return 1  # Low
    elif value <= tertiles[1]:
        return 2  # Medium
    else:
        return 3  # High

def classify_mobility(value, tertiles):
    if value <= tertiles[0]:
        return 1  # Low
    elif value <= tertiles[1]:
        return 2  # Medium
    else:
        return 3  # High

# Apply classification to both variables
new_data['student_class'] = new_data['Observation'].apply(
    lambda x: classify_students(x, student_tertiles))
new_data['mobility_class'] = new_data['chn2021'].apply(
    lambda x: classify_mobility(x, mobility_tertiles))

# Create combined bivariate class: 0 for no students, 1â€“9 for all other combinations
# This enables plotting with a 3x3 bivariate color scheme
new_data['bivariate_class'] = new_data.apply(
    lambda row: 0 if row['student_class'] == 0 else
    row['student_class'] + (row['mobility_class'] - 1) * 3,
    axis=1)
```

```python


# Define bivariate color scheme - include special color for zero-value areas
# Define colors: special color for no students, then the 3x3 bivariate scheme
bivariate_colors = [
    '#f7f7f7',  # No students - Very light gray (almost white)
    '#0063D1',  # Low students-Low mobility - Very light blue
    '#FFD6A5',  # Medium students-Low mobility - Light blue
    '#FFAA5E',  # High students-Low mobility - Medium blue
    '#2785FF',  # Low students-Medium mobility - Very light yellow
    '#FFF4E6',  # Medium students-Medium mobility - Light blue-gray
    '#FF8427',  # High students-Medium mobility - Blue-green
    '#5EB0FF',  # Low students-High mobility - Orange
    '#A5D8FF',  # Medium students-High mobility - Medium blue
    '#D96400'   # High students-High mobility
]

# Map categories to colors
color_dict = {i: color for i, color in enumerate(bivariate_colors)}

# Draw the bivariate map
fig, ax = plt.subplots(figsize=(15, 12))
new_data.plot(column='bivariate_class',
                categorical=True,
                cmap=colors.ListedColormap(bivariate_colors),
                linewidth=0.3,
                edgecolor='#333333',
                ax=ax)

# Add hatching pattern for areas with no students
no_data_areas = new_data[new_data['student_class'] == 0]
no_data_areas.plot(ax=ax, facecolor="none", edgecolor='gray',
                     hatch='///', linewidth=1, alpha=0.9)

# Create a grid for the 3x3 bivariate legend
legend_ax = fig.add_axes([0.15, 0.19, 0.15, 0.15])
legend_ax.set_xlim(0, 3)
legend_ax.set_ylim(0, 3)

# Add color squares
for i in range(3):
    for j in range(3):
        idx = i * 3 + j + 1  # +1 because we have a special category at index 0
        legend_ax.add_patch(plt.Rectangle((j, i), 1, 1, facecolor=bivariate_colors[idx], edgecolor='white'))

# Remove legend axes
legend_ax.set_xticks([])
legend_ax.set_yticks([])

# Add legend labels
legend_ax.text(1.5, -0.45, 'Economically Inactive Students', ha='center', va='center', fontsize=10)
legend_ax.text(-0.45, 1.5, 'Residential Mobility Index', ha='center', va='center', rotation=90, fontsize=10)

# Add directional arrows
legend_ax.text(0.5, -0.2, 'Low', ha='center', va='center', fontsize=8)
legend_ax.text(2.5, -0.2, 'High', ha='center', va='center', fontsize=8)
legend_ax.text(-0.2, 0.5, 'Low', ha='center', va='center', fontsize=8)
legend_ax.text(-0.2, 2.5, 'High', ha='center', va='center', fontsize=8)

# Add a special legend for areas with no students
no_data_patch = mpatches.Patch(facecolor=bivariate_colors[0], edgecolor='gray', hatch='///', label='No Data')
no_data_legend_ax = fig.add_axes([0.11, 0.36, 0.15, 0.02])
no_data_legend_ax.axis('off')
no_data_legend=no_data_legend_ax.legend(handles=[no_data_patch], loc='center' ,
    frameon=True,
    framealpha=0.9,
    title='Special Areas')

# Add title and annotations
ax.set_title('Spatial relationship between Economically Inactive Students and Residential Mobility Index', fontsize=20, loc='center')

plt.figtext(0.12, 0.1,
           "Data source: Census 2021 https://www.ons.gov.uk/census/maps/ & Consumer Data Research Centre https://data.cdrc.ac.uk/dataset/cdrc-residential-mobility-index",
           fontsize=10)
ax.set_axis_off()
add_north_arrow(ax=ax, scale=.75, xlim_pos=0.1, ylim_pos=.85, color='#000', text_scaler=2, text_yT=-1.25)

# Add a scale bar
scalebar = ScaleBar(
    1, "m",
    location="lower left"
)
ax.add_artist(scalebar)

plt.show()
```

The non-spatial visualization above supports the conclusion that economically inactive students are more strongly affected by the residential mobility index in Leeds. The merged dataset must be cleaned to explore the spatial relationship between economically inactive students and the residential mobility index. First, only the data, including economically inactive students, was required to be retained, and this data was joined with the LSOA shapefile for Leeds through the LSOA code. The newly merged dataset exploration revealed that 50 LSOAs have missing values for both economically inactive student data and residential mobility data. In order to maintain the authenticity of the spatial visualization in the next step, these data were retained, and these special areas were marked during the visualization process. This spatial visualization applies a bivariate choropleth map to illustrate the spatial relationship between the two variables. This map can efficiently show the spatial relationship between two variables by dividing each variable's data into three categories and using a combination matrix legend with nine colors to facilitate displaying the associations between variables [10]. Before the spatial visualization, the newly merged dataset needs to be processed by classifying both student numbers and the residential mobility index into categories using tertiles, and then appending the classified data to the table. This classification method is less influenced by extreme values and has relatively high accuracy, making it suitable for the data in this study [11]. The classified data were calculated to classify each LSOA into nine classes. Additionally, missing values were assigned a zero value to ensure they could be identified and marked on the map. The final classified data were used for spatial visualization, and to enhance accessibility and ensure readability for all audiences, a high-contrast color scheme consisting of five shades of orange and four shades of blue was adopted [12]. According to the map, areas with high residential mobility index and large economically inactive student populations are primarily concentrated in the central-western part of Leeds, with similar patterns also observed in some northeastern and eastern coastal areas. Conversely, areas with lower residential mobility index and fewer students are mainly located in the east part of Leeds and some areas of the southwest and northwest.
